```python
import pandas as pd
import pycountry
import numpy as np
```

```python
eonet = pd.read_json("../eonet.json")
disaster_data = pd.read_csv("../Preprocessed-Natural-Disasters.csv", delimiter=";")
```

```python
eonet["iso"] = eonet.apply(lambda row: list(row["geometries"])[0]["country"], axis=1)
eonet["date"] = eonet.apply(lambda row: list(row["geometries"])[0]["date"], axis=1)
eonet["latitude"] = eonet.apply(lambda row: list(row["geometries"])[0]["coordinates"][0], axis=1)
eonet["longitude"] = eonet.apply(lambda row: list(row["geometries"])[0]["coordinates"][1], axis=1)
eonet["year"] = pd.to_datetime(eonet["date"]).dt.year
eonet["day"] = pd.to_datetime(eonet["date"]).dt.day
eonet["month"] = pd.to_datetime(eonet["date"]).dt.month
```

```python
eonet = eonet[eonet["iso"] != ""]
eonet = eonet[~eonet["iso"].isnull()]
eonet["iso"] = eonet["iso"].dropna()
eonet["iso"] = eonet.apply(lambda row: pycountry.countries.get(alpha_2=row["iso"]).alpha_3, axis=1)
```

```python
def merge(row):
    events = eonet[eonet["iso"] == row["ISO"]]
    for _, event in events.iterrows():
        if row["Start Year"] == event["year"]:
            if row["Start Month"] == event["month"]:
                if row["Start Day"] == event["day"]:
                    row["Latitude"] = event["latitude"]
                    row["Longitude"] = event["longitude"]
                    print(row)
    return row
```

```python
disaster_data[disaster_data["Latitude"].isna() | disaster_data["Longitude"].isna()] = disaster_data[disaster_data["Latitude"].isna() | disaster_data["Longitude"].isna()].apply(lambda row: merge(row), axis=1)
disaster_data
```
